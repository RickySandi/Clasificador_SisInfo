```python
import pandas as pd
import numpy as np
import nltk

import string
from nltk.tokenize import sent_tokenize
from bs4 import BeautifulSoup
```

```python
#data.sample(frac=0.1).shape
#pd.DataFrame()


data_prueba=data.sample(frac=0.01)
palabrasParada = set( nltk.corpus.stopwords.words('english') + list(string.punctuation))

```

```python
def borrarSiContiene(palabra, borrar):
    if borrar in palabra:
        return 
    return palabra
def Tokenizar(texto):
    texto=texto.lower()
    sinHTML=BeautifulSoup(texto).text
    #sinHTML=remove_html_tags(texto)#quita html
    palabras = nltk.word_tokenize(sinHTML)#separa las palabras
    return [palabra for palabra in palabras if palabra not in palabrasParada]#quita stopwords
```

```python
def listToString(s):  
    str1 = " " 
    return (str1.join(s)) 

listaDeComentarios=[]
for index, row in data_prueba.iterrows():
    a=Tokenizar(row['Review'])
    x=listToString(a)  
    z={'oracion':x,'clasificacion':row['Label'],'original':row['Review']}
    listaDeComentarios.append(z)

```

```python
import collections

for i in listaDeComentarios:
    count_by_token = collections.Counter(i.get("oracion").split())
    print([key for key, _ in count_by_token.most_common(4)],i.get("clasificacion"))


```

```python

for a in listaDeComentarios:
    a['contados'] = collections.Counter(a.get("oracion").split())
    
    
```

```python
from nltk.corpus import wordnet as wn
ListaDeAdjetivos=[]
for synset in list(wn.all_synsets("a")):
    a = (str(synset))
    result= (a.split("'")[1].split(".")[0])
    if(result not in ListaDeAdjetivos):
        ListaDeAdjetivos.append(result)
        
((ListaDeAdjetivos))#lista de adjetivos sin repetir  
```

```python
listaDeComentarios[0]
```

```python
for i in listaDeComentarios:
    i["palabrasAdjetivas"]=[]
    listaDePalabrasSinRepetir=list(i["contados"])
    for x in listaDePalabrasSinRepetir:
        if(x in ListaDeAdjetivos):
            i["palabrasAdjetivas"].append(x)
listaDeComentarios[0]
```

```python
paraEntrenar={}
count=0
for i in listaDeComentarios:
    count =count+1
    paraEntrenar[str(count)]=[i["clasificacion"],i["palabrasAdjetivas"]]
    
paraEntrenar
```
